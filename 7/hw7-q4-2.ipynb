```julia
using JuMP, Cbc

N = 1:6
a = Dict(zip(N,[8, 11, 6, 5, 7, 5]))
P = [(1,3),(2,4),(3,4),(3,5),(4,5),(4,6)]
c = 4000
f = 10000

m = Model(Cbc.Optimizer)

@variable(m, x[N]>= 0)
@variable(m,b[N], Bin)
@variable(m,z>=0)
@variable(m, y[N] >= 0)


@objective(m, Min, sum(4000 * y[i] for i in N) + sum(10000 * b[i] for i in N))

@constraint(m, z <= 18)
@constraint(m,max_time[i in N],z>= x[i]+a[i]-y[i]) 
@constraint(m,task_start[(i,j) in P],x[j]>= x[i]+a[i]-y[i]) 

for i in N
    @constraint(m, y[i] <= 4 * b[i])
end



optimize!(m)

println("Minimum total cost: ", objective_value(m))
for i in N
    println("Activity ", i, ": Start time = ", value(x[i]), ", Duration reduction = ", value(y[i]))
end
println("Project makespan: ", value(z))
```
